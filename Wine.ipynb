```python
data=np.array([[14.23],[13.20],[13.16],[14.37],[13.24]])
```

```python
linked=linkage(data,method='single')
```

```python
plt.figure(figsize=(8,5))
```

```python
dendrogram(linked,orientation='top',labels=[14.23,13.20,13.6,14.37,13.24],distance_sort='ascending',show_leaf_counts=True,
           truncate_mode='lastp')
plt.title('wine Clustering Dendrogram')
plt.xlabel('sample index or cluster Size')
plt.ylabel('distance')
plt.grid(True)
plt.show()
```
